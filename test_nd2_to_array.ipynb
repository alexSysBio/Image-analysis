### Example of using the nd2_to_array function

<br> In this example an nd2 file with no iteration axis is used, since it is small in size. This is why the iteraction axis is shown as '', the channels list is empty and the number of time-points or positions are zero.

<br> This function has been tested for the following iterations axes:
1. ''
2. 't'
3. 'mc'
4. 'mct'

<br> It should work for the remaining iterations as well. If not please reach-out and I will implement any exceptions.

```python
# importing libraries and path to the directory where the functions are stored
import sys
sys.path.append('/Users/alexandros/Documents/Visual_code/Image-analysis')
import nd2_to_array as ndtwo
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
# define the image path
image_path = '.../test_bead_image_zero_iter.nd2'
```

```python
# run the nd2 reader
image_read = ndtwo.nd2_to_array(image_path)
```

```python
# check the returned results
print('Iteration axis')
print(image_read[0], 'this interaction axis is zero.')
print('nd2 fields')
print(image_read[1])
print('Show image')
plt.imshow(image_read[2], cmap='gray', vmin=50, vmax=2000)
plt.xticks([])
plt.yticks([])
plt.show()
print('Channels list')
print(image_read[3])
print('Number of timepoints')
print(image_read[4])
print('Number of positions')
print(image_read[5])
print('Scale ("um/px")')
print(image_read[6])
print('Sensor size (px)')
print(image_read[7])
```
