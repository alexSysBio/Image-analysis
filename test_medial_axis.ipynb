### Example for running the get_medial_axis function included in the Biviriate_medial_axis_estimation.py script

<br> This notebook demonstrates how the biviriate medial axis estimation is performed for elongated $Escherichia$ $coli$ cells ($dnaC2$ conditional mutant).
<br> An exemplary set of parameters is used.

@author: Alexandros Papagiannakis, HHMI at Stanford University, Jan 11 2025

Cite: https://www.biorxiv.org/content/10.1101/2024.10.08.617237v2.full

    DNA/polysome phase separation and cell width confinement couple nucleoid segregation 
    to cell growth in Escherichia coli

    Alexandros Papagiannakis, Qiwei Yu, Sander K. Govers, Wei-Hsiang Lin,  Ned S. Wingreen, Christine Jacobs-Wagner

    bioRxiv, https://doi.org/10.1101/2024.10.08.617237, October 22, 2024

```python
# Import libraries and the path where the Biviriate_medial_axis_estimation.py script is stored
import sys
sys.path.append('.../Image-analysis')
import Biviriate_medial_axis_estimation as bma
import pickle
```

```python
# load the test dictionary that contains the cell IDs as keys, and the cropped cells masks as values
with open ('.../medial_axis_test_cell_masks', 'rb') as handle:
     mask_dict = pickle.load(handle)
cropped_cell_mask = mask_dict[list(mask_dict.keys())[0]]

```

```python
# iterate to the first 10 cell masks and run the get_medial_axis and get_oned_coordinate functions
for i in range(10):
    cropped_cell_mask = mask_dict[list(mask_dict.keys())[i]]

    medial_axis = bma.get_medial_axis(cropped_cell_mask, radius_px=8, half_angle=22, cap_knot=13, max_degree=60, verbose=True)
    oned_axis  = bma.get_oned_coordinates(cropped_cell_mask, medial_axis[0])
    print(oned_axis)
```
